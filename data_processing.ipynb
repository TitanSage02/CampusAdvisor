```python
import os
from docx import Document

def convert_docx_to_txt(input_directory, output_directory):
    # Crer le dossier de sortie s'il n'existe pas
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # Parcourir tous les fichiers dans le dossier d'entre
    for filename in os.listdir(input_directory):
        if filename.endswith(".docx"):
            filepath = os.path.join(input_directory, filename)

            # Ouvrir et lire le contenu du fichier docx
            document = Document(filepath)
            full_text = []
            for paragraph in document.paragraphs:
                full_text.append(paragraph.text)

            # Crer le contenu du fichier .txt
            txt_content = "\n".join(full_text)

            # Dfinir le chemin du fichier .txt de sortie
            txt_filename = os.path.splitext(filename)[0] + ".txt"
            txt_filepath = os.path.join(output_directory, txt_filename)

            # crire le contenu dans le fichier .txt
            with open(txt_filepath, "w", encoding="utf-8") as txt_file:
                txt_file.write(txt_content)

# Exemple d'utilisation
input_directory = "data_word"
output_directory = "data"
convert_docx_to_txt(input_directory, output_directory)

print("Conversion termine.")
```
